```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt

from math import radians, cos, sin, asin, sqrt
%matplotlib inline
```

```python
origin = pd.DataFrame()
origin['Latitude'] = np.arange(0,91,5.0)
origin['Longitude'] = 0.0
origin['Type'] = 'Origin'
```

```python
origin
```

```python
vertical = pd.DataFrame()
vertical['Latitude'] = origin['Latitude'] - 0.1
vertical['Longitude'] = origin['Longitude']
vertical['Type'] = 'Vertical'
```

```python
horizontal = pd.DataFrame()
horizontal['Latitude'] = origin['Latitude']
horizontal['Longitude'] = origin['Longitude'] - 0.1
horizontal['Type'] = 'Horizontal'
```

```python
coordinates = pd.concat([origin, vertical, horizontal], ignore_index=True)
```

```python
coordinates
```

```python
for point_type, group in coordinates.groupby('Type'):
    plt.scatter(group['Longitude'], group['Latitude'], label=point_type)

# Add labels and legend
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Coordinates by type')
plt.legend()
plt.show()
```

```python
sb.scatterplot(data = coordinates, x = 'Longitude', y = 'Latitude', hue = 'Point')
```
